## Data Visualization

***Bar Plot***

```python
# 
sns.barplot(data = df, x = 'Category', y = 'Sales')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Har Category Ka Total Sales')
```

```python
# 
sns.barplot(data = df, x = 'Region', y = 'Profit')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.title('Har Region Ka Total Profit')
```

***Scatter Plot***

```python
sns.scatterplot(data = df, x = 'Sales', y = 'Profit', hue = 'Category')
```

```python
sns.scatterplot(data = df, x = 'Quantity', y = 'Sales', hue = 'Category')
plt.xlabel('Quantity')
plt.ylabel('Sales')
plt.title('Har Category ka Quantity and Sales')
```

***Line Plot***

```python
sns.lineplot(data = df, x = 'Quantity', y = 'Sales', hue = 'Region')
plt.xlabel('Quantity')
plt.ylabel('Sales')
plt.title('Har Region Ki Quantity And Sales')
```

```python
sns.lineplot(data = df, x = 'Discount', y = 'Sales', hue = 'Category')
plt.title('Har Category ka Discount and Sales')
```

***Pie Chart***

```python
df1 = df['Ship Mode'].value_counts()
plt.pie(df1.values, labels = df1.index, autopct = '%1.1f%%')
plt.title('Ship Mode Distribution')
```

```python
df5 = df['Region'].value_counts()
plt.pie(df5.values, labels = df5.index, autopct='%1.1f%%')
plt.title('Orders Distribution by Region')
```

***Box Plot***

```python
sns.boxplot(data = df, x = 'Region', y = 'Profit', hue = 'Discount')
plt.title('Region aur Discount ke hisab se Profit Distribution')
```

***Correlation Heatmap***

```python
df2 = df[['Sales','Quantity','Discount','Profit']].corr()
sns.heatmap(df2, annot = True, cmap = 'coolwarm')
plt.title('Numeric Features Correlation Matrix')
```

***Bar Plot***

```python
sns.barplot(data = df, x = 'Segment', y = 'Sales', hue = 'Region')
plt.title('Customer Segment ke hisab se Average Sales')
plt.ylabel('Average Sales')
```

***Violin Plot***

```python
sns.violinplot(data = df, x = 'Quantity', y = 'Sales', hue = 'Category')
plt.title('Quantity Ke Hisab Se Sales Distribution')
```
