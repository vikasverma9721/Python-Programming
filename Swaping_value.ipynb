```python
## Without using third variable
def swap_function_without_using_third_variable(a,b):
    a = input("Enter the first value =")
    b = input("Enter the second value =")
    print("Befor swapping value :","a =",a,",","b =",b)
    a,b = b,a
    print("After swapping value :","a =",a,",","b =",b)
```

```python
swap_function_without_using_third_variable(a,b)
```

```python
## With using third variable
def swap(a,b):
    a = input("Enter the first value")
    b = input("Enter the second value")
    print("Befor swapping value :","a =",a,",","b =",b)
    temp = a
    a = b
    b = temp
    print("After swapping value :","a =",a,",","b =",b)
```

```python
swap(a,b)
```
