## Exploratory Data Analysis (EDA) for Iris Dataset

```python
# Import requried libries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = sns.load_dataset('iris')
```

## Data Cleaning

```python
# Show head of your data
df.head()
```

```python
# Show info of our data
df.info()
```

```python
# Describe your data
Describe_data = df.describe()
print(Describe_data)
```

```python
# Find missing values of your data
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Find Duplicate values of your data
Duplicate_values = df.duplicated()
print(Duplicate_values)
```

## Data Visualization

***Scatter Plot***

```python
# Displays the features of each species (such as sepal length vs sepal width) in different colors.
sns.scatterplot(data = df, x = 'sepal_length', y = 'sepal_width', hue = 'species')
```

***Pair Plot***

```python
# Shows the relationship between all possible feature pairs in a single figure.
sns.pairplot(data = df, hue = 'species')
```

***Histogram***

```python
# Displays the frequency distribution of a single feature
sns.histplot(data = df, x = 'sepal_length', hue = 'species', multiple = 'stack')
```

***Box Plot***

```python
# Shows the distribution of feature values and identifies outliers
sns.boxplot(data = df, x = 'species', y = 'sepal_length')
```

***Violin Plot***

```python
# Combines the features of both box plot and histogram.
sns.violinplot(data = df, x = 'species', y = 'sepal_length')
```

***Heatmap***

```python
# Displays the correlation between features.
df1 = df.select_dtypes(include = ['number'])
plt.figure(figsize = (8,6))
sns.heatmap(df1.corr(), annot = True, cmap = 'coolwarm')
```

***Bar Plot***

```python
sns.barplot(data = df)
```

```python
# Shows the count or average valuesofspecies.
sns.barplot(data = df, x = 'species', y = 'sepal_length')
```

```python
sns.lineplot(data = df, x = 'sepal_length', y = 'sepal_width')
```

```python
sns.countplot(data = df)
```
