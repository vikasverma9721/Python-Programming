***List = A list in Python is a collection of items that ordered, changeable, and allow duplicate elements.
List are one of the most versatile data structures in Python and used to store multiple items in a single variable.***

***Creating a List***

```python
# create a list
list1 = [1,2,3,4,5,6,4,5,9,4,7,5,98]  # List of integers
print(list1)  
```

***Check the data type***

```python
type(list1)
```

***Accessing Value***

***Change Item Value***

```python
l1 = [45,25,41,26,35,78,89,45]
print(l1)
```

***Append Method***

```python
# Append
list1 = [1,2,3,4,5,6,7]  # list of integers
list2 = ['apple','banana','grapes']  # list of string
list1.append(list2)
print(list1)
```

***Remove Method***

```python
# Remove method
list1 = ['apple','banana','orange','grapes']  # list of string
list1.remove("banana")
print(list1)
```

```python
num = [1,5,4,7,8,6,2]   # list of integers
num.remove(5)
print(num)
```

***Insert Method***

```python
# Insert method
color = ['Red','Green','Yellow','Pink']  # list of string
color.insert(2,'White')
print(color)
```

```python
x = ['Alica','Alexa','Charlie','Zohan']  # list of string
x.insert(1,'Lencho')
print(x)
```

***Extend Method***

```python
# Extend method
students = ['Alica','Alexa','Charlie','Zohan','Marry']  # list of string
courses = ['BSc','BCA','BA','MCA','AI']  # list of string
student.extend(courses)
print(student)
```

```python
list1 = ['A','B','C','D','E']  # list of string
list2 = [1,2,3,4,5]  # list of integers
list1.extend(list2)
print(list1)
```

***POP Method***

```python
friuts = ['Apple','Banana','Papaya','Cherry']  # list of string
friuts.pop(1)
print(friuts)
```

```python
x = ['a',1,'b',3,'c',4,'d',5,'e']  # mixed 
x.pop(2)
print(x)
```

***Delete Method***

```python
name = ['Alica','Alexa','Charlie','Zohan','Marry']  # list of string
del name
```

```python
print(name)
```

***Clear Method***

```python
color = ['Red','Green','Yellow','Pink']  # list of string
color.clear()
print(color)
```

```python
x = ['a',1,'b',3,'c',4,'d',5,'e']  # mixed 
x.clear()
print(x)
```
