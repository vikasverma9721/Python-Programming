## Data Visualization

***Content Volume and Types***

```python
# Count of movies vs TV shows
sns.countplot(x='type', data=df) #Shows whether Netflix leans more toward movies or series in this dataset
```

```python
# Highlights growth in content production and which years/decades have the most releases.
sns.countplot(x='year', data=df)
```

***Ratings and Popularity***

```python
# Reveals how wellâ€‘rated Netflix content is overall and whether ratings are skewed toward high or mediocre scores
sns.histplot(df['rating'], kde=True)
```

***Genre and country analysis***

```python
#Finds the most common genres (e.g., Drama, Comedy, Documentary) in the catalog
df['genres'].str.split(',').explode().str.strip().value_counts().head(10).plot.bar()
```

```python
#Shows which countries contribute the most content to Netflix
df['country'].str.split(',').explode().str.strip().value_counts().head(10).plot.bar()
```

***Audience / age certification***

```python
#Reveals whether the library leans toward family content (G/PG) or mature content (16+, 18+)
sns.countplot(x='rating', data=df, order=sorted(df['rating'].dropna().unique()))
```

***Multiâ€‘feature / advanced plots***

```python
num_cols = ['rating','imdb_votes','tmdb_popularity','runtime']  # keep only those that exist
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm')
```
