## Data Visualization

***Dis Plot***

```python
sns.displot(data = df)
```

***Pair Plot***

```python
sns.pairplot(data = df)
```

```python
sns.kdeplot(data = df)
```

***Scatter Plot***

```python
sns.scatterplot(data = df, x = 'total_bill', y = 'tip')
```

```python
sns.scatterplot(data = df, x = 'day', y = 'tip')
```

```python
sns.relplot(data = df, x = 'total_bill', y = 'tip', kind = 'scatter')
```

```python
sns.lineplot(data = df, x = 'total_bill', y = 'tip')
```
