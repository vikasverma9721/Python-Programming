## Data Visualization

***Distribution Plot***

```python
# Visualizes the distribution of house prices.
sns.displot(df5['price'])
```

```python
sns.displot(df5['bedrooms'])
```

***Correlation Heatmap***

```python
# 1. Select only numerical columns (excludes 'date' and 'street')
numeric_df = df5.select_dtypes(include=['number'])

# 2. Calculate the correlation matrix
correlation_matrix = numeric_df.corr()

# 3. Plot the heatmap
plt.figure(figsize=(12, 10)) # Adjust size so it's readable
sns.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm')
plt.title('Correlation Heatmap for House Prices')
```

***Scatter Plot***

```python
# Visualization relationship between a feature and price.
plt.scatter(df5['bathrooms'], df5['price'])
plt.xlabel('bathrooms')
plt.ylabel('price')
plt.title('bathrooms vs Price')
```

***Pair Plot***

```python
# Shows pairwise relationships between all numerical features.
sns.pairplot(df5)
```

***Residaul Plot***

```python
# Shows residuals (actual - predicted) for model evaluation.
sns.residplot(x = df5['bathrooms'], y = df5['price'])
plt.title('Residaul Plot')
```

***Box Plot***

```python
# Visualizes price distribution across categories (e.g., location).
sns.boxplot(x ='bedrooms', y ='price', data = df5)
plt.title('Price Distribution by bedrooms')
```

***Line Plot***

```python
# Plots price trends over time.
plt.plot(df5['bedrooms'], df5['price'])
plt.xlabel('date')
plt.ylabel('price')
plt.title('House Price Over Time')
```

```python
sns.lineplot(data = df5, x = 'bedrooms', y = 'price')
plt.xlabel('bedrooms')
plt.ylabel('price')
plt.title('Line Plot')
```

***Hist Plot***

```python
sns.histplot(data = df5, x = 'floors')
```

***Kde Plot***

```python
sns.kdeplot(data = df5, x = 'bedrooms')
```
