## While Loop.

```python
# while loop we can execute a set of statement as long as a condition is true.
i = 1
while i < 6:
    print(i)
    i = i + 2
```

```python
x = 2
while x < 14:
    print(x)
    x = x + 1
```

```python
# Print even Numbers from 0 to 10
i = 0
while i < 10:
    print(i)
    i = i + 2
```

```python
# Print odd Numbers from 1 to 20
i = 1
while i < 20:
    print(i)
    i = i + 2
```

```python
# Break statement we can stop the loop even if the while condition is true.
i = 1
while i < 10:
    print(i)
    if i == 4:
        break
    i += 1
```

```python
x = 0
while i < 20:
    print(i)
    if i == 18:
        break
    i += 2
```

```python
# Continue statement we can stop the loop even if the while condition is true.
i = 0
while i < 10:
    i += 1
    if i == 11:
        continue
    print(i)
```

```python
# WAP to find sum of first n natural numbers.
n = int(input("Enter the last number:"))
i = 1
sum = 0
while (i <= n):
    sum = sum + i
    i = i + 1
print("The sum is: ",sum)
```
