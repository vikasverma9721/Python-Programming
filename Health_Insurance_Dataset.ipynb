```python
plt.figure(figsize=(6,4))
sns.countplot(data = df, x = "smoker", hue = "smoker")
plt.show()
```

```python
plt.figure(figsize = (6,4))
sns.countplot(data = df, x = "smoker", hue = "sex")
plt.title("Count of male smokers vs female smokers")
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.boxplot(data = df, x = "smoker", y = "charges")
plt.title("Is Insurence charges depanded on smokers")
plt.show()
```

```python
plt.figure(figsize = (8,5))
sns.scatterplot(data = df, x = "bmi", y = "charges", hue = "smoker")
plt.title("Charges depends on BMI")
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.scatterplot(data = df, x = "age", y = "charges")
plt.title("Do charges increase with Age")
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.scatterplot(data = df, x = "age", y = "charges", hue = "smoker")
plt.title("Do charges increase with age & smoker")
plt.show()
```

```python
plt.figure(figsize = (6,4))
sns.barplot(data = df, x = "sex", y = "charges")
plt.title("Do charges depends on sex")
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.barplot(data = df, x = "region", y = "charges", hue = "region")
plt.title("In which region is treatment most expensive")
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.heatmap(df.corr (numeric_only = True), annot = True, cmap = "coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
